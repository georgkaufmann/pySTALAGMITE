# Dissolving calcite in water enriched with carbon dioxide
In this notebook, we review the **dissolution** of calcite (CaCO$_3$) along the walls of a fracture,
both in **open-** and **closed-system** conditions.

We will use the library`libCHEM.py` for the reaction constants.

```python
import numpy as np
import matplotlib.pyplot as plt
import libCHEM
```

----
# Calcite dissolution
Calcite can be dissolved in a fracture by agressive water flowing through. The reaction is fairly
slow, depends on the carbon-dioxide pressure (either **open** to the atmosphere, or **closed** to the atmosphere),
and can be described by three **rate-limiting** reactions.
<table><tr>
<td><img src="images/chemical_reactions_open.jpg" style=width:80mm></td>
<td><img src="images/chemical_reactions_closed.jpg" style=width:80mm></td>
</tr></table>

$$
\begin{array}{rcl}
  {\rm CaCO}_3 + {\rm H}^+             
  & \overset{\kappa_1}{\underset{\kappa'_4}{\leftrightarrow}} & {\rm Ca}^{2+} + {\rm HCO}_3^- \\
  {\rm CaCO}_3 + {\rm H}_2{\rm CO}_3^0 
  & \overset{\kappa_2}{\underset{\kappa''_4}{\leftrightarrow}} & {\rm Ca}^{2+} + 2{\rm HCO}_3^- \\
  {\rm CaCO}_3 + {\rm H}_2{\rm O}      
  & \overset{\kappa_3}{\underset{\kappa'''_4}{\leftrightarrow}} & {\rm Ca}^{2+} + {\rm CO}_3^{2-} 
\end{array}
$$

Here, the $\kappa_i$ are rate coefficients.

- The first process is dissolution due to the presence of H$^+$ in a strongly acidic environment, 
- The second process is a direct reaction of calcite with the carbonic acid H$_2$CO$_3$, 
- The third process describes the physical dissolution of calcite in water. 

For pH-values larger than six, which is true for natural karst waters, the second reaction is negligible. 
For CO$_2$-pressures smaller than 0:01 atm, the third reaction is the dominant reaction. 

The calcite dissolution as a **fairly slow process** is also rate-limiting.

```python
TC = np.linspace(0,200,101)

fig,axs = plt.subplots(2,2,figsize=(8,8))
axs[0,0].set_xlim(0,200)
axs[0,0].set_xlabel('Temperature $T$ [$^{\circ}$C]')
axs[0,0].set_ylabel('$\kappa_1$ [m/s]')
axs[0,0].semilogy(TC,libCHEM.PWP(TC)[0],linewidth=3,color='red')
axs[0,0].grid()
axs[0,1].set_xlim(0,200)
axs[0,1].set_xlabel('Temperature $T$ [$^{\circ}$C]')
axs[0,1].set_ylabel('$\kappa_2$ [m/s]')
axs[0,1].semilogy(TC,libCHEM.PWP(TC)[1],linewidth=3,color='red')
axs[0,1].grid()
axs[1,0].set_xlim(0,200)
axs[1,0].set_xlabel('Temperature $T$ [$^{\circ}$C]')
axs[1,0].set_ylabel('$\kappa_3$ [mol/m$^2$/s]')
axs[1,0].semilogy(TC,libCHEM.PWP(TC)[2],linewidth=3,color='red')
axs[1,0].grid()
axs[1,1].set_xlim(0,200)
axs[1,1].set_xlabel('Temperature $T$ [$^{\circ}$C]')
axs[1,1].set_ylabel('$\kappa_4$ [m$^4$/mol/s]')
axs[1,1].semilogy(TC,libCHEM.PWP(TC)[3],linewidth=3,color='red',label='pco2=0.00042 atm')
axs[1,1].semilogy(TC,libCHEM.PWP(TC,pco2=0.01)[3],linewidth=3,color='red',linestyle='dashed',label='pco2=0.001 atm')
axs[1,1].semilogy(TC,libCHEM.PWP(TC,pco2=0.1)[3],linewidth=3,color='red',linestyle='dotted',label='pco2=0.1 atm')
axs[1,1].grid()
axs[1,1].legend()
plt.tight_layout()
plt.savefig('images/CHEM_kappa.png')
```

If the calcite dissolution occurs under **equilibrium conditions**, we can summarize the three rate-limiting equations 
to the **solubility product** $K_C$ [mol$^2$/l$^2$] at equilibrium is given by:
$$
K_C = (\textrm{Ca}^{2+})(\textrm{CO}_3)
$$
with the activities given in round brackets.

```python
TC = np.linspace(0,200,101)

plt.figure(figsize=(8,4))
plt.title('Solubility product calcite')
plt.xlabel('Temperature $T$ [$^{\circ}$C]')
plt.ylabel('$K_C$ [mol$^2$/l$^2$]')
plt.plot(TC,libCHEM.KC(TC),color='blue',linestyle='solid',linewidth=2,label='$K_C$')
plt.plot(TC,libCHEM.KC(TC,S=10),color='blue',linestyle='dashed',linewidth=2,label='$K_C, S=10$')
plt.plot(TC,libCHEM.KC(TC,S=35),color='blue',linestyle='dotted',linewidth=2,label='$K_C, S=35$')
plt.yscale('log')
plt.grid()
plt.legend(loc='lower left',shadow=True)
plt.savefig('images/CHEM_KC.png')
```

## Calcium equilibrium concentration

Next, we consider a sloution **saturated with respect to calcite**.

We recall the definition of the **ionic strength**. The ionic strength of a solution is defined as
$$
  I = \frac{1}{2} \sum\limits_i z_i^2 [X_i]
$$
Here, $z_i$ is the ionic charge, and $[X_i]$ (in mol l$^{-1}$) the concentration of species $X_i$.

Using the dissolved species Ca$^{2+}$, H$^+$, OH$^-$, HCO$_3^-$, CO$_3^{2-}$, find:
$$
I = \frac{1}{2} 
\left( 4 [{\rm Ca}^{2+}] + [{\rm H}^{+}] + [{\rm OH}^{-}] + 4 [{\rm CO}_3^{2-}] + [{\rm HCO}_3^{-}] \right)
$$

The **equation of electro-neutrality**, when neglecting ion pairs, reads:
$$
2 [{\rm Ca}^{2+}] + [{\rm H}^+] = [{\rm HCO}_3^-] + 2 [{\rm CO}_3^{2-}] + [{\rm OH}^-]
$$

In natural karst water, with $4<pH<8$ and neglecting ion pairs, the concentration of hydrogen,
hydroxyl, and carbonate can be neglected.
Then ionic strength and electro-neutrality relax to 
$$\begin{array}{rcl}
I &\simeq& 3 [{\rm Ca}^{2+}] \\
2 [{\rm Ca}^{2+}] &=& [{\rm HCO}_3^-]
\end{array}$$

For solutions, which are saturated with respect to calcite, a simplified form of the charge balance
can be used: $2 [{\rm Ca}^{2+}] = [{\rm HCO}_3^-]$. 

### Open system
Using these equations, and observing the simplified charge balance, after
some algebraic manipulations we obtain an analytical expression for the calcium equilibrium concentration
as a function of carbon-dioxide partial pressure:
$$\fbox{$
 [{\rm Ca}^{2+}]^3_{eq} = \frac{K_1 K_C K_H}{4 K_2 
 \gamma_{{\rm Ca}^{2+}} \gamma^2_{{\rm HCO}_3^{2+}} } p_{CO_2}
$}$$
This relation is valid for the open system.

```python
pco2 = np.zeros([0,0])
ceq1o = np.zeros([0,0])
ceq2o = np.zeros([0,0])
ceq3o = np.zeros([0,0])
ceq4o = np.zeros([0,0])
for p in np.arange(400,50000,step=10):
    pco2 = np.append(pco2,p)
    temp = 5.
    ceq1o = np.append(ceq1o,libCHEM.CEQ_limestone_open(temp,p/1.e6))
    temp = 10.
    ceq2o = np.append(ceq2o,libCHEM.CEQ_limestone_open(temp,p/1.e6))
    temp = 15.
    ceq3o = np.append(ceq3o,libCHEM.CEQ_limestone_open(temp,p/1.e6))
    temp = 20.
    ceq4o = np.append(ceq4o,libCHEM.CEQ_limestone_open(temp,p/1.e6))
   #print(temp,ceq_open)
```

```python
plt.figure(figsize=(8,4))
plt.title('Calcium equilibrium concentration (open system)')
plt.xlim([0,4])
plt.ylim([300,60000])
plt.xlabel('Calcium concentration $c_{eq}$ [mol/m$^3$]')
plt.ylabel('Carbon-dioxide pressure $p$ [ppm]')
plt.semilogy(ceq1o,pco2,color='green',linestyle='solid',linewidth=3,label='T=5$^{\circ}$')
plt.semilogy(ceq2o,pco2,color='blue',linestyle='solid',linewidth=3,label='T=10$^{\circ}$')
plt.semilogy(ceq3o,pco2,color='orange',linestyle='solid',linewidth=3,label='T=15$^{\circ}$')
plt.semilogy(ceq4o,pco2,color='red',linestyle='solid',linewidth=3,label='T=20$^{\circ}$')
plt.legend(loc='upper left',shadow=True)
plt.grid()
plt.savefig('images/CHEM_CEQ_PCO2_open.png')
```

### Closed system
For the closed system, the carbon dioxide is consumed during dissolution, and the carbon-dioxide pressure will reduce
according to:
$$
 p_{CO_2} = p_{CO_2}^{initial} - \frac{[{\rm Ca}^{2+}]_{eq}}{K_H (1 + {1 \over {K_0}})}, 
 \quad p_{CO_2}^{initial} > 7 \times 10^{-4} \mbox{atm}
$$

$$\fbox{$
 [{\rm Ca}^{2+}]^3_{eq} = \frac{K_1 K_C K_H}{4 K_2 
 \gamma_{{\rm Ca}^{2+}} \gamma^2_{{\rm HCO}_3^{2+}} } p_{CO_2}
$}$$

```python
pco2 = np.zeros([0,0])
ceq1c = np.zeros([0,0])
ceq2c = np.zeros([0,0])
ceq3c = np.zeros([0,0])
ceq4c = np.zeros([0,0])
for p in np.arange(400,50000,step=10):
    pco2 = np.append(pco2,p)
    temp = 5.
    ceq1c = np.append(ceq1c,libCHEM.CEQ_limestone_closed(temp,p/1.e6))
    temp = 10.
    ceq2c = np.append(ceq2c,libCHEM.CEQ_limestone_closed(temp,p/1.e6))
    temp = 15.
    ceq3c = np.append(ceq3c,libCHEM.CEQ_limestone_closed(temp,p/1.e6))
    temp = 20.
    ceq4c = np.append(ceq4c,libCHEM.CEQ_limestone_closed(temp,p/1.e6))
   #print(temp,ceq_open)
```

```python
plt.figure(figsize=(8,4))
plt.title('Calcium equilibrium concentration (closed system)')
plt.xlim([0,3])
plt.ylim([300,60000])
plt.xlabel('Calcium concentration $c_{eq}$ [mol/m$^3$]')
plt.ylabel('Carbon-dioxide pressure $p$ [ppm]')
plt.semilogy(ceq1o,pco2,color='green',linestyle='solid',lw=1)
plt.semilogy(ceq2o,pco2,color='blue',linestyle='solid',lw=1)
plt.semilogy(ceq3o,pco2,color='orange',linestyle='solid',lw=1)
plt.semilogy(ceq4o,pco2,color='red',linestyle='solid',lw=1,label='open')

plt.semilogy(ceq1c,pco2,color='green',linestyle='solid',linewidth=3,label='T=5$^{\circ}$')
plt.semilogy(ceq2c,pco2,color='blue',linestyle='solid',linewidth=3,label='T=10$^{\circ}$')
plt.semilogy(ceq3c,pco2,color='orange',linestyle='solid',linewidth=3,label='T=15$^{\circ}$')
plt.semilogy(ceq4c,pco2,color='red',linestyle='solid',linewidth=3,label='T=20$^{\circ}$')
plt.legend(loc='lower right',shadow=True)
plt.grid()
plt.savefig('images/CHEM_CEQ_PCO2_open.png')
```
